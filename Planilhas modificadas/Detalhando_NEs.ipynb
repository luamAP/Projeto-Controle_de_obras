```python
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.action_chains import ActionChains

from time import sleep as para
from time import time as contar

class Navegacao():
    def __init__(self):
        self.navegador = webdriver.Chrome()
        self.vars = {'ignored_exceptions': (NoSuchElementException,StaleElementReferenceException), 'senha': ['Jaspekde27!','Jaspekde27.9', 'Jaspekde279310','Luam27.9','Luam2722']}

    def esperar_elemento(self, param1, param2, elementos=False, tempo_parar=60):
        """Espera um elemento carregar na página

        Args:
            param1 (): tipo do elemento
            param2 (str): nome do elemento
            tempo0 (int, optional): tempo determinado para executar a espera. Padrão 120 (2 min).
        """    
        tempo0 = contar() 
        print(f'Esperando elemento ({param2})')
        if elementos:
            print('Entra no "if elemento:"', end=' ')
            try:
                print('"try:"', end=' ')
                elemento =self.navegador.find_elements(param1, param2)
                while ((tempo-tempo0)>=tempo_parar) and (len(elemento)==0):
                    print('"while elemento:"', end=' ')
                    para(1)
                    tempo = contar() 
                para(1)
                if tempo>tempo0:
                    print('"if tempo>tempo0:"', end=' ')
                    print(f'Tempo excedido!')
                else:
                    print('"else: (tempo>tempo0)"', end=' ')

            except:
                print('"except:"', end=' ')
                elemento = WebDriverWait(self.navegador, tempo_parar,ignored_exceptions=self.vars['ignored_exceptions'])\
                            .until(expected_conditions.presence_of_all_elements_located((param1, param2)))
                tempo = contar()

            finally:
                print('"finally:"')
                para(1)
                print(f'>>> Elementos "{param2}" encontrados em {tempo-tempo0}s! (len({len(elemento)}))')
                return elemento
        else:
            print('Entra no "else:"', end=' ')
            try:
                print('"try:"', end=' ')
                elemento =self.navegador.find_elements(param1, param2)
                while ((tempo-tempo0)>=tempo_parar) and (len(elemento)==0):
                    print('"while elemento:"', end=' ')
                    para(1)
                    tempo = contar() 
                para(1)
                if tempo>tempo0:
                    print('"if tempo>tempo0:"', end=' ')
                    print(f'Tempo excedido!')
                else:
                    print('"else: (tempo>tempo0)"', end=' ')
                    
            except:
                print('"except:"', end=' ')
                elemento = WebDriverWait(self.navegador, tempo_parar, ignored_exceptions=self.vars['ignored_exceptions'])\
                                .until(expected_conditions.presence_of_element_located((param1, param2)))
                tempo = contar()
            
            finally:
                print('"finally:"')
                para(1)
                try:
                    print(f'>>> Elemento "{elemento.text}" encontrado em {tempo-tempo0}s!')
                except:
                    print(f'>>> Elemento "{elemento}" encontrado em {tempo-tempo0}s!')
                
                return elemento

    ### Esperar uma página carregar -> *_esperar_aba(param)_*
    def esperar_aba(self):
        print('Esperando aba/página abrir!')
        paginas = len(self.navegador.window_handles)
        paginas_update = len(self.navegador.window_handles)

        while paginas == paginas_update:
            paginas_update = len(self.navegador.window_handles)
            para(1)
        para(1)
        
        print('Página carregada!')

    def AFIM_PesquisaNE(self, afim):
        """Abre o AFIM e direciona para a função LISNE

        Args:
            afim (_type_): Ano no qual o FIM irá abrir
        """
        print(f'\n=============== Executando entra_AFIM {afim}! ===============')
        
        if afim in ['2014','2015','2016','2017','2018']:
            #AFIM DE 2018
            if afim=='2018': self.navegador.get('http://afim2.manaus.am.gov.br/AfimPRD2018/')
            #AFIM DE 2017
            elif afim=='2017': self.navegador.get('http://afim2.manaus.am.gov.br/AfimPRD2017/')
            #AFIM DE 2016
            elif afim=='2016': self.navegador.get('http://afim2.manaus.am.gov.br/AfimPRD2016/')
            #AFIM DE 2015
            elif afim=='2015': self.navegador.get('http://afim2.manaus.am.gov.br/AfimPRD2015/')
            #AFIM DE 2014
            elif afim=='2014': self.navegador.get('http://afim2.manaus.am.gov.br/AfimPRD2014/')
        
            self.navegador.find_element(By.NAME, 'user').send_keys('03441042220')
            self.navegador.find_element(By.NAME, 'password').send_keys(self.vars['senha'][3])
            self.navegador.find_element(By.NAME, 'enviar').click()
            self.navegador.switch_to.window(self.navegador.window_handles[-1])
            self.navegador.switch_to.frame(1)

            print(f'=============== Finalizando entra_AFIM {afim}! ===============\n')
            print('Selecionando a opção LISNE!')

            para(2)

            try:
                self.navegador.find_element(By.CSS_SELECTOR, "body").click()
                self.navegador.find_element(By.XPATH, "//li[contains(.,\'EXEORC\')]").click()
                self.navegador.find_element(By.XPATH, "(//li[@id=\'foldheader\'])[7]").click()
                self.navegador.find_element(By.LINK_TEXT, "LISNE").click()
                print('Entrando no Try:')

            except:
                self.navegador.esperar_elemento(By.CSS_SELECTOR, "body").click()
                self.navegador.esperar_elemento(By.XPATH, "//li[contains(.,\'EXEORC\')]").click()
                self.navegador.esperar_elemento(By.XPATH, "(//li[@id=\'foldheader\'])[7]").click()
                self.navegador.esperar_elemento(By.LINK_TEXT, "LISNE").click()
                print('Entrando no Except:')

        #AFIM DE 2019
        if afim =='2019':
            self.navegador.get('https://afim1.manaus.am.gov.br/AfimPRD2019/')
            self.navegador.find_element(By.ID, 'user').send_keys('034.410.422-20')
            # Mudar as senhas periodocamente
            self.navegador.find_element(By.ID, 'password').send_keys(self.vars['senha'][3])
            self.navegador.find_element(By.ID, 'password').send_keys(Keys.ENTER)

            para(2)
            for w in self.navegador.window_handles:
                self.navegador.switch_to.window(w)
                if 'mensagem' in self.navegador.current_url: self.navegador.close()
                elif afim in self.navegador.current_url: pass
                else: self.navegador.close()
                
            self.navegador.switch_to.window(self.navegador.window_handles[-1])
            self.navegador.switch_to.frame(1)

            try:
                self.navegador.find_element(By.CSS_SELECTOR, "body").click()
                self.navegador.find_element(By.XPATH, "//li[contains(.,\'EXEORC\')]").click()
                self.navegador.find_element(By.XPATH, "(//li[@id=\'foldheader\'])[7]").click()
                self.navegador.find_element(By.LINK_TEXT, "LISNE").click()
                print('Entrando no Try:')

            except:
                self.navegador.esperar_elemento(By.CSS_SELECTOR, "body").click()
                self.navegador.esperar_elemento(By.XPATH, "//li[contains(.,\'EXEORC\')]").click()
                self.navegador.esperar_elemento(By.XPATH, "(//li[@id=\'foldheader\'])[7]").click()
                self.navegador.esperar_elemento(By.LINK_TEXT, "LISNE").click()
                print('Entrando no Except:')

        if afim in ['2020', '2021', '2022']:
            #AFIM DE 2022
            if afim=='2022': self.navegador.get('https://afim1.manaus.am.gov.br/AfimPRD2022/')
            #AFIM DE 2021
            elif afim=='2021': self.navegador.get('https://afim1.manaus.am.gov.br/AfimPRD2021/')
            #AFIM DE 2020
            elif afim=='2020': self.navegador.get('https://afim1.manaus.am.gov.br/AfimPRD2020/')
            
            self.navegador.find_element(By.ID, 'user').send_keys('034.410.422-20')
            # Mudar as senhas periodocamente
            self.navegador.find_element(By.ID, 'password').send_keys(self.vars['senha'][3])
            self.navegador.find_element(By.ID, 'password').send_keys(Keys.ENTER)
            self.navegador.switch_to.window(self.navegador.window_handles[-1])
            self.navegador.switch_to.frame(0)

            try:
                self.navegador.find_element(By.CSS_SELECTOR, "body").click()
                self.navegador.find_element(By.XPATH, "//li[contains(.,\'EXEORC\')]").click()
                self.navegador.find_element(By.XPATH, "(//li[@id=\'foldheader\'])[7]").click()
                self.navegador.find_element(By.LINK_TEXT, "LISNE").click()
                print('Entrando no Try:')

            except:
                self.navegador.esperar_elemento(By.CSS_SELECTOR, "body").click()
                self.navegador.esperar_elemento(By.XPATH, "//li[contains(.,\'EXEORC\')]").click()
                self.navegador.esperar_elemento(By.XPATH, "(//li[@id=\'foldheader\'])[7]").click()
                self.navegador.esperar_elemento(By.LINK_TEXT, "LISNE").click()

        ### Mudando de janela,
        # self.esperar_aba()

    def LISNE_detalhada(self, NE):
        """Pesquisa a Nota de Empenho desejada para retirar as informações

        Args:
            NE (_type_): Nota de Empenho

        Returns:
            _type_: Tupla com as informações extraidas
        """
        print(f'\n=============== Executando LISNE! ({NE}) ===============')
        para(2)
        self.navegador.switch_to.window(self.navegador.window_handles[-1])

        self.esperar_elemento(By.NAME, 'cboUnidadeGestora').send_keys('270101 - Secretaria Municipal de Infraestrutura')
        self.esperar_elemento(By.ID, 'mkdNumeroNE').clear()
        self.esperar_elemento(By.ID, 'mkdNumeroNE').send_keys(NE)

        try:
            self.esperar_elemento(By.ID, 'IMGprocurar').click()

        except:
            self.esperar_elemento(By.ID, 'mkdNumeroNE').send_keys(Keys.ENTER)

        para(2)
        
        elemento = self.esperar_elemento(By.CSS_SELECTOR, 'tr td.readCell')
        ActionChains(self.navegador)\
        .double_click(elemento)\
        .perform()

        para(1)

        self.navegador.switch_to.window(self.navegador.window_handles[-1])
        
        credor_empresa = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtCredor"]').get_attribute('value')
        evento = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtEvento"]').get_attribute('value')
        convenio = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtConvenio"]').get_attribute('value')
        programa_de_trabalho = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtPT"]').get_attribute('value')
        unidade_orcamentaria = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtUnidadeOrcamentaria"]').get_attribute('value')
        fonte_de_recurso = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtFonte"]').get_attribute('value')
        natureza_despesa = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtNaturezaDespesa"]').get_attribute('value')
        modalidade = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtModalidade"]').get_attribute('value')
        valor_total = self.esperar_elemento(By.CSS_SELECTOR, 'tr td input[name="txtValor"]').get_attribute('value')
        descricao = self.esperar_elemento(By.CSS_SELECTOR, 'tr td textarea[name="txtDescricao"]').get_attribute('value')
        
        # valor_01 = self.esperar_elemento(By.CSS_SELECTOR, 'tr td [name="txtCronograma1"]').get_attribute('value')
        
        self.navegador.close()
        
        return (credor_empresa, evento, convenio, programa_de_trabalho, unidade_orcamentaria, fonte_de_recurso, natureza_despesa, modalidade, valor_total, descricao)
```

```python
from openpyxl import load_workbook, Workbook

def formatar_NE(caminho, file):
    """Formatando as Notas de Empenho para o padrão, 20XXNE00XXXX

    Args:
        caminho (_type_): Pasta onde se encontra o arquivo
        file (_type_): Arquivo onde os dados serão processados

    Returns:
        _type_: Arquivo salvo com as informações processadas
    """

    pl = load_workbook(caminho+'\\'+file)
    file = file.split('.')
    aba_ativa = pl.active
    for celula in aba_ativa["AB"]:
        # print(celula.value)
        if None == celula.value: pass
        elif '.' in celula.value or '/' in celula.value:
            l = celula.row
            if '.' in celula.value: cell = celula.value[1:]
            else: cell = celula.value
            # print(cell, cell.split(' '))
                
            for n in cell.split(' '):
                # print(n)
                if '/' in n:
                    cell_list = n.split('/')
                    if len(cell_list[1]) > 2: 
                        cell_list[1] = cell_list[1][2:]
                    cell_final = '20'+cell_list[1]+'NE'+cell_list[0]

            aba_ativa[f'AB{l}'] = cell_final

    novo_arq = f'{file[0]}(formatado).{file[1]}'
    pl.save(caminho+'\\'+novo_arq)
    print(f'Salvando a planilha -> {novo_arq}')
    
    return novo_arq

def por_ano(ano,file):
    """_summary_

    Args:
        ano (_type_): _description_
        file (_type_): _description_
        T (_type_): Instância da Classe Navegacao

    Returns:
        _type_: _description_
    """

    T = Navegacao()
    pl = load_workbook(file)
    aba_ativa = pl.active
    nova_aba = pl.create_sheet('Aba teste NEs detalhadas')
    T.AFIM_PesquisaNE(ano)
    
    for celula in aba_ativa["AB"]:
        l = celula.row
        if None == celula.value: pass
        elif 'NE' in celula.value:
            cell = celula.value.split('NE')
            if cell[0]==ano:
                dados = T.LISNE_detalhada(cell[1])

                aba_ativa[f'AC{l}'] = dados[0]
                aba_ativa[f'AD{l}'] = dados[1]
                aba_ativa[f'AE{l}'] = dados[2]
                aba_ativa[f'AF{l}'] = dados[3]

    pl.save(file)
    
    windows = T.navegador.window_handles

    T.navegador.switch_to.window(windows[-1])
    T.navegador.close()
    T.navegador.switch_to.window(windows[0])

    return file

def formatando_numeros(file, ano_corrente):
    pl = load_workbook(file)
    aba_ativa = pl.active

    for letra in ['C','D','E','F']:
        for celula in aba_ativa[f"A{letra}"]:
            l = celula.row
            if None == celula.value: pass
            else:
                c = celula.value.split('-')
                aba_ativa[f"A{letra}{l}"] = c[0]

    file = f'CONTROLE DE OBRAS-{ano_corrente}(finalizado).xlsx'
    pl.save(file)
    return file

def criando_cabecalho(file, linha_NEs):
    pl = load_workbook(file)
    aba_ativa = pl.active
    nova_aba = pl.create_sheet('Aba teste NEs detalhadas')

    nova_aba['A1'] = 'NOTA DE EMPENHO'
    nova_aba['B1'] = 'CREDOR'
    nova_aba['C1'] = 'EVENTO'
    nova_aba['D1'] = 'CONVÊNIO'
    nova_aba['E1'] = 'PROGRAMA DE TRABALHO'
    nova_aba['F1'] = 'UNIDADE ORÇAMENTÁRIA'
    nova_aba['G1'] = 'FONTE DE RECURSO'
    nova_aba['H1'] = 'NATUREZA DE DESPESA'
    nova_aba['I1'] = 'MODALIDADE'
    nova_aba['J1'] = 'VALOR'
    nova_aba['K1'] = 'DESCRIÇÃO'

    linha = 1

    for celula in aba_ativa[linha_NEs]:
            if None == celula.value: pass
            elif 'NE' in celula.value:
                linha += 1
                nova_aba[f'A{linha}'] = celula.value

    pl.save(file)

```
